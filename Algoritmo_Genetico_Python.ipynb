```python
import random
import numpy as np
```

```python
T = 60                    #periodo
tamPopulacao = 2          #quatidade de individuos da população 
rotas = 3                 #quatidade de rotas
tipoVeiculos = 3          #tipos de veiculos
qntdVeiculos = [12,8,14]  # ft_b - veiculos disponiveis para cada tipo


#iniciando uma população aleatoria
populacao = []  

for i in range(tamPopulacao): 
    individuo = np.zeros((tipoVeiculos, rotas))
    for b in range(tipoVeiculos):
        for r in range(rotas):
            individuo[b][r] = random.randint(0, qntdVeiculos[b])
    populacao.append(individuo)
```

```python
#imprimindo a população
print("\n __Populacao aleatoria__")
for individuo in populacao:
    print(individuo, end="\n")
print("_"*20)


veiculoEnviado = []
viavel = True

K = [100, 45, 67]
S = 10

for i in range(tamPopulacao):
    print("Individuo ",(i+1))
    individuo = populacao[i] 
    for b in range(tipoVeiculos):
        soma=0
        for r in range(rotas):  
            soma = soma +  individuo[b][r]
        print("Foram enviados ", soma, "veiculos do tipo", b)

        H_br = np.zeros((tipoVeiculos, rotas))

        #checar se a soma de veiculos enviados nao ultrapassa a frota
        if soma > qntdVeiculos[b]:
            viavel=False
            break
        else:
            viavel=True
            for b in range(tipoVeiculos):
                for r in range(rotas):
                    if individuo[b][r]==0:
                        H_br[b][r]=0
                    else:
                        H_br[b][r] = T/individuo[b][r]
    print()
    if viavel:
        veiculoEnviado.append(H_br)
        #imprimindo o H_br
        for i in veiculoEnviado:
            print(i, end="\n")

        #custo operacional 
        co=0
        for b in range(tipoVeiculos):
            for r in range(rotas):
                co = co + (K[b] * (individuo[b][r]))

        cs=0                              
        #custo social
        for b in range(tipoVeiculos):
            for r in range(rotas):
                cs = cs + (S * (H_br[b][r]))               
                
        #fitness
        fitness = co + cs
        print("\nFitness: ", fitness)

    print("\n")
    
```
